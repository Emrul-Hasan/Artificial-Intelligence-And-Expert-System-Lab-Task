# Table driven agent

```python
Thistable={('A','Light'):'On',('A','Dark'):'Down',('B','Light'):'Off',('B','Dark'):'Up'}
percepts1=[]    
def table_driven_agent(percept1):
    print(percept1)
    percepts1.append(percept1)
    action = lookup(percept1,Thistable)
    return action
def lookup(percept1,Thistable):
    for key in Thistable:
        if key == percept1:
            return Thistable[key]
import random
Location = random.choice(['A','B'])
Condition = random.choice(['Light','Dark'])
while True:
    action= table_driven_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'):break
           
    if action == 'On':
        Location = 'B'
        Condition = random.choice(['Light','Dark'])
    elif action== 'Off':
        Location = 'A'
        Condition = random.choice(['Light','Dark'])
    else:
        Condition = 'Light'
```
