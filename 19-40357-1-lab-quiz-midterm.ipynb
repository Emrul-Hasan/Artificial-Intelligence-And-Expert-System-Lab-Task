```python

table={('A','Clean'):'Go C',('A','Dirty'):'Suck',('B','Clean'):'Go A',('B','Dirty'):'Suck', ('C','Clean'):'Go B',('C','Dirty'):'Suck'}
percepts=[]    
def table_driven_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action

def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]

import random
Location = random.choice(['A','B','C'])
Condition = random.choice(['Clean','Dirty'])

while True:
    action= table_driven_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Go B':
        Location = 'B'
        Condition = random.choice(['Clean','Dirty'])
    elif action== 'Go C':
        Location = 'C'
        Condition = random.choice(['Clean','Dirty'])
    elif action== 'Go A':
        Location = 'A'
        Condition = random.choice(['Clean','Dirty'])
    else:
        Condition = 'Clean'   
```
