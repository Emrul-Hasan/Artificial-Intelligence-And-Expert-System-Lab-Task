```python
h_bucharest = {'Arad': 366,'Zerind': 374,'Oradea': 380,'Sibiu': 253,
'Timisoara': 329,'Lugoj': 244,'Mehadia': 241,'Drobeta': 242,
'Craiova': 160,'Rimnicu': 193,'Fagaras': 176,'Pitesti': 100,
'Bucharest': 0,'Giurgiu': 77,'Urziceni': 80,'Hirsova': 151,
'Eforie': 161,'Vaslui': 199,'Iasi': 226,'Neamt': 234
}
```

```python
from queue import PriorityQueue


def greedybfs(startingNode, destinationNode):
    visited = {}    #to store explored nodes
    distance = {}
    parent = {}   #to store path

    bestfirstsearch_traversal_output = []     #to store the order of traversal
    frontier = PriorityQueue()

    for city in romanian_map.keys():
        visited[city] = False
        parent[city] = None
        distance[city] = -1

    startingCity = startingNode
    visited[startingCity] = True
    distance[startingCity] = 0
    frontier.put((h_bucharest[startingCity],startingCity))    # oradea

    while not frontier.empty():
        u = frontier.get()[1]     # in the first iteration it will be arad
        bestfirstsearch_traversal_output.append(u)
        if u == destinationNode:
            break
        visited[u] = True
        for v in romanian_map[u].keys():
            if not visited[v]:                
                parent[v] = u
                distance[v] = distance[u] + romanian_map[u][v]                
                frontier.put((h_bucharest[v],v))
                
            
    g = destinationNode
    path = []
    while g is not None:
        path.append(g)
        g = parent[g]
    
    path.reverse()
    print(path)
    print(distance[destinationNode])
    print(bestfirstsearch_traversal_output)

greedybfs('Oradea', 'Bucharest')
```

# A* search 

```python
from queue import PriorityQueue


def AstarAlgo(startingNode, destinationNode):
    visited = {}
    distance = {}
    parent = {}

    AstarAlgo_traversal_output = []
    frontier = PriorityQueue()

    for city in romanian_map.keys():
        visited[city] = False
        parent[city] = None
        distance[city] = -1

    startingCity = startingNode
    visited[startingCity] = True
    distance[startingCity] = 0
    frontier.put((h_bucharest[startingCity],startingCity))

    while not frontier.empty():
        u = frontier.get()[1]     
        AstarAlgo_traversal_output.append(u)
        if u== destinationNode:
            break
        visited[u] = True
        for v in romanian_map[u].keys():
            if not visited[v]:                
                parent[v] = u
                distance[v] = distance[u] + romanian_map[u][v]
                frontier.put((distance[v],v))

    g = destinationNode
    path = []
    while g is not None:
        path.append(g)
        g = parent[g]

    path.reverse()
    print(path)
    print(distance[destinationNode])
    print(AstarAlgo_traversal_output)

AstarAlgo('Oradea', 'Bucharest')
```
