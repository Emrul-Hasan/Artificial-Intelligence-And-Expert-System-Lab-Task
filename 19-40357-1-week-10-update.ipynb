```python
# row attack
def rowAttack(r,c,N):
    for i in range(0,N):
        for j in range(i+1,N):
            if(r[i] == r[j]):
                print("\n")
                print('Row attack: row:{} column:{} and {}'.format(r[i],c[i],c[j]))   
                return 0
    return 1    
```

```python
#column attack
def columnAttack(r,c,N):
    for i in range(0,N):
        for j in range(i+1,N):
            if(c[i] == c[j]):
                print("\n")
                print('Column attack: row:{} and {} column:{} '.format(r[i],r[j],c[j]))      
                return 0
    return 1
```

```python
# diagonal atack
def diagonalAttack(r,c,N):
    for i in range(0,N):
        for j in range(i+1,N):
            if(abs(c[i]-c[j]) == abs(r[i]-r[j])):
                print("\n")
                print('Diagonal attack: position:{},{} and {},{}  '.format(c[i],r[i],c[j],r[j]))  
                return 0
    return 1
```
